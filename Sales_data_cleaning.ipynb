```python
import pandas as pd
```

```python
df = pd.read_csv(r"E:\Recap-Data Analysis\Python\cleaning and exploration\Clean\dirty_business_sales.csv")
df
```

```python
df.columns = [c.strip().lower().replace(" ",'_') for c in df.columns]
```

```python
a = df['sale_date'].astype(str).str.strip()
b = pd.to_datetime(a,errors='coerce')
for fmt in ['%Y-%m-%d','%Y/%m/%d','%m/%d/%Y','%d/%m/%Y']:
    c = b.isna()
    if not c.any():
        break
    b.loc[c] = pd.to_datetime(a[c],errors='coerce',format=fmt)

df['sale_date'] = b.dt.strftime('%Y-%m-%d')
```

```python
df
```

```python
for c in ['store_id','product_id','category','customer_email']:
    if c in df.columns:
        df[c] = df[c].astype(str).str.strip()
        df[c] = df[c].str.title()

df
```

```python
for c in ['quantity','unit_price','discount','revenue']:
    if c in df.columns:
        df[c] = df[c].astype(str).str.replace("[^0-9.\-]",'',regex=True)
        df[c] = pd.to_numeric(df[c],errors='coerce')

df
```

```python
for c in ['discount']:
    if c in df.columns:
        d = df[c] > 1
        df.loc[d,c] = df.loc[d,c]/100.0

df
```

```python
df['customer_email'] = df['customer_email'].str.replace(".Example",'@Example')
```

```python
df
```

```python
df['customer_email'] = df['customer_email'].str.replace("Ex Ample",'Example')
```

```python
df
```

```python
df = df.fillna('')
```

```python
df
```

```python
df = df.replace("Nan",'')
```

```python
df
```

```python
for x in df.index:
    if df.loc[x,'sale_date'] =='':
        df.drop(x,inplace=True)
```

```python
for x in df.index:
    if df.loc[x,'quantity'] =='':
        df.drop(x,inplace=True)
```

```python
for x in df.index:
    if df.loc[x,'unit_price'] =='':
        df.drop(x,inplace=True)
```

```python
for x in df.index:
    if df.loc[x,'discount'] =='':
        df.drop(x,inplace=True)
```

```python
for x in df.index:
    if df.loc[x,'revenue'] =='':
        df.drop(x,inplace=True)
```
